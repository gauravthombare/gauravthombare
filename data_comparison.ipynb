```python
from openpyxl import load_workbook
import pandas as pd
import difflib
```

# File information

```python
filepath=r"/content/drive/MyDrive/cauvery_calling/GP_Wise_mapping_comparison.xlsx"
```

# Using Pandas

```python
xls=pd.ExcelFile(filepath)
sheet_names=xls.sheet_names
print(sheet_names)
user_sheet=sheet_names[0]
master_sheet=sheet_names[1]
diff_sheet=sheet_names[2]
```

Sheet Updated by User

```python
separator="-"
```

```python
df1 = pd.read_excel(xls, user_sheet)
df1=df1.drop(['Sl.NO','Remarks'],axis=1)
```

```python
df2 = pd.read_excel(xls, master_sheet)
df2 = df2.set_axis(["District_Code", "District_Name","Taluk_Code","Taluk_Name","GP_Code","GP_Name"], axis=1)
```

```python
# Not using the codes data to compare because interested in only GP data
# df2['unique_data']=df2["District_Code"].astype(str)+separator+df2["District_Name"].str.lower()\
# +separator+df2["Taluk_Code"].astype(str)+separator+df2["Taluk_Name"].str.lower()\
# +separator+df2["GP_Code"].astype(str)+separator+df2["GP_Name"].str.lower()

df2['unique_data']=df2["District_Name"].str.lower()+separator+df2["Taluk_Name"].str.lower()+separator+df2["GP_Name"].str.lower()
```

Label Correction

```python
num_rows=len(df1)
for i in range(num_rows):
  try:
    df1['District_Name'][i]=difflib.get_close_matches(df1['District_Name'][i].upper(),df2['District_Name'].str.upper())[0]
  except:
      pass
  try:
    df1['Taluk_Name'][i]=difflib.get_close_matches(df1['Taluk_Name'][i].upper(),df2['Taluk_Name'].str.upper())[0]
  except:
    pass
    try:
      df1['GP_Name'][i]=difflib.get_close_matches(df1['GP_Name'][i].upper(),df2['GP_Name'].str.upper())[0]
    except:
      pass
```

```python

df1['unique_data']=df1["District_Name"].str.lower()+separator+df1["Taluk_Name"].str.lower()+separator+df1["GP_Name"].str.lower()
```
